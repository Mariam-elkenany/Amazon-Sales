```python
Sales['currency'] = Sales['currency'].fillna('EGP')  # Replace nulls in currency with 'EGP'
```

```python
Sales['Courier Status'] = Sales['Courier Status'].fillna('delivered')  # Replace nulls in Courier Status with 'delivered'
```

```python
Sales['UnitPrice'] = Sales['UnitPrice'].fillna(0)  # Replace nulls in UnitPrice with 0 for the orders cancelled 
```

```python
Sales['ship-country'] = Sales['ship-country'].fillna('EG')  # Replace nulls in ship-country with 'EG'
```

```python
Sales['fulfilled-by'] = Sales['fulfilled-by'].fillna('FBA')  # Replace nulls in fulfilled-by with 'FBA' fullfilled by Amazon
```

```python
Sales['promotion-ids'] = Sales['promotion-ids'].fillna('no promotions applied')  # Replace nulls for the orders that does not have discount
```

```python
Sales['B2B'] = Sales['B2B'].fillna('FALSE')
```

```python
Sales.isna().sum()
```

```python
sales_by_fulfilment = Sales.groupby('Fulfilment')['SubTotal'].sum().reset_index()

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x='Fulfilment', y='SubTotal', data=sales_by_fulfilment)
plt.title('Total Sales by Fulfilment Status')
plt.xlabel('Fulfilment Status')
plt.ylabel('Total Sales (SubTotal)')

plt.show()
```

```python
# Merge inventory data with warehouse information
inventory_warehouse = pd.merge(Inventories, Warehouses, left_on='WarehouseID', right_on='warehouseID')

# Group data by 'warehouseName' and calculate total quantity
inventory_by_warehouse = inventory_warehouse.groupby('warehouseName')['OnShelfQuantity'].sum().reset_index()

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x='warehouseName', y='OnShelfQuantity', data=inventory_by_warehouse)
plt.title('Total On-Shelf Inventory by Warehouse')
plt.xlabel('Warehouse Name')
plt.ylabel('Total On-Shelf Quantity')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(Product['Price'], bins=20, kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
merged_data = pd.merge(Sales, Product, on='ProductID')

# 1. Horizontal Bar Chart of Total Sales by Category
# Group by category and sum the SubTotal for each category
total_sales_by_category = merged_data.groupby('Category')['SubTotal'].sum().reset_index()

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=total_sales_by_category, x='SubTotal', y='Category', palette='viridis')
plt.title('Total Sales by Category (Horizontal Bar Chart)')
plt.xlabel('Total Sales ($)')
plt.ylabel('Category')
plt.show()
```

```python
Sales['Date'] = pd.to_datetime(Sales['Date'])

# Group by date and sum the SubTotal
sales_over_time = Sales.groupby('Date')['SubTotal'].sum().reset_index()

# Create a line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=sales_over_time, x='Date', y='SubTotal', marker='o')
plt.title('Sales Over Time (Line Chart)')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
